*   BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
*   BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
*   BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
*   BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
*   BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
*   PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
*   PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
*   PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
*   PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
*   PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
*   PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
*   default.payment.next.month: Default payment (1=yes, 0=no)

Não possui valores faltantes. Todas as variáveis estão marcadas como inteiro, porém se olharmos o significado delas, temos dois grupos distintos: variáveis númericas e variáveis categóricas.

É importante destacarmos essa diferença entre as variáveis pois algumas análises estatísticas, como o boxplot por exemplo, não faz muito sentido para variáveis categóricas. Para esse tipo de variável existem outros tipos de análise que acabam sendo mais eficazes.

# Análise Exploratória

A análise explatória será dividida em três: uma só para as variáveis numéricas, outra para variáveis categóricas e uma análise entre as duas categorias. A razão disso é tornar a análise mais adequada para cada tipo de variável. Por exemplo, um histograma não faz muito sentido para a variável SEX, mas um gráfico de waffle sim. Para a variável AGE um gráfico de barras não passaria uma informação clara sobre a variável, um boxplot nesse caso seria melhor.

```python
# dataset de variáveis numéricas
dados_numer = dados[['AGE', 'LIMIT_BAL', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 
                     'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 
                     'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']]

# dataset de variáveis categóricas
dados_cat = dados[['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 
                   'PAY_4', 'PAY_5', 'PAY_6', 'DEFAULT_PAY']]
```

Para observar melhor os valores reais dos limitantes inferiores e superiores, vamos gerar uma tabela.

```python
# Exibindo valores dos limites inferiores e superiores de cada variável numérica
def get_box_plot_data(labels, bp):
    rows_list = []

    for i in range(len(labels)):
        dict1 = {}
        dict1['Variável'] = labels[i]
        dict1['Limite inferior'] = bp['whiskers'][i*2].get_ydata()[1]
        dict1['Q1'] = bp['boxes'][i].get_ydata()[1]
        dict1['Q2 - Mediana'] = bp['medians'][i].get_ydata()[1]
        dict1['Q3'] = bp['boxes'][i].get_ydata()[2]
        dict1['Limite superior'] = bp['whiskers'][(i*2)+1].get_ydata()[1]
        rows_list.append(dict1)

    return pd.DataFrame(rows_list)

get_box_plot_data(['AGE', 'LIMIT_BAL', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 
                     'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 
                     'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'], 
                  plt.boxplot([dados_numer['AGE'], dados_numer['LIMIT_BAL'],
                               dados_numer['BILL_AMT1'], dados_numer['BILL_AMT2'], 
                               dados_numer['BILL_AMT3'], dados_numer['BILL_AMT4'], 
                               dados_numer['BILL_AMT5'], dados_numer['BILL_AMT6'], 
                               dados_numer['PAY_AMT1'], dados_numer['PAY_AMT2'], 
                               dados_numer['PAY_AMT3'], dados_numer['PAY_AMT4'], 
                               dados_numer['PAY_AMT5'], dados_numer['PAY_AMT6']
                              ],widths=0)
                 )
```

Observando a tabela gerada com as informações vindas dos boxplots, vemos que:

AGE: Confirmamos que os idosos (clientes acima de 60 anos) são outliers.

LIMIT_BAL: Antes estimado valores acima de 550.000 como outliers, agora confirmamos que valores acima de 520.000 mil já se enquadram como outliers.

BILL_AMT1 ... BILL_AMT6: Antes dito que valores acima de 400.000 são bem raros, aqui observamos que os valores por volta de 150.000 já são considerados outliers. Da mesma forma, antes dito que os valores negativos são raros, agora vemos que para BILL_AMT1 valores abaixo de -15.000 e para BILL_AMT2...BILL_AMT6 por volta de -60.000 são considerados outliers.

PAY_AMT1 ... PAY_AMT6: Antes dito que valores acima de 50.000 poderiam ser outliers, aqui vemos que valores por volta de 10.000 já são considerados outliers.

#### Análise de correlação

Quando dizemos que duas variáveis são correlacionadas entre si, queremos dizer que ambas possuem uma relação. No caso, essa relação pode ser positiva ou negativa.
Em uma correlação positiva temos que quando o valor de uma variável aumenta, o valor de uma outra variável também aumenta. Na correlação negativa temos que quando o valor de uma variável aumenta, o valor de uma outra variável diminui.
O valor do índice de correlação vai de -1 até 1, onde -1 a correlação entre as variáveis é perfeitamente negativa, e o valor 1 indica que as variáveis são correlacionadas de forma perfeita e positiva. Além disso, correlações que apresentam valor menor que 0.39, ou -0.39, nos diz que as variáveis são correlacionadas de forma fraca. Logo, só iremos considerar as correlações que forem maiores que 0.40, ou -0.40.

## Impacto financeiro

Queremos maximizar os verdadeiro positivo e minimizar os falso negativo.

Recall alto, porém temos R$50,00 por ligação, então a precisão seria boa para não gastar muito nisso.

Se o cliente pagar, teremos um lucro de 5% do valor na fatura, se não pagar, perdemos tudo. 

Se não fizermos nada, conseguimos recuperar 30% do valor atrasado de toda a carteira.

Queremos saber quanto a empresa deixará de perder e a taxa de inadimplência.

Aprofundando melhor sobre o valor pendente a ser pago, entendo que:

Eu tenho uma fatura de X e pago Y;
No mês seguinte eu tenho que pagar X-Y + X', sendo X' minha nova fatura e eu pago Y';
No mês seguinte eu tenho que pagar X+X'-Y-Y' + X'' e eu pago Y'';
E assim por diante.
Em cima disso, posso ter ou não meses de atraso no pagamento.

Portanto, utilizaremos essa conta da soma entre as faturas e subtração entre os valores pagos para determinar o impacto financeiro.

Consideraramos que essa soma sob os clientes marcados como default é o prejuízo se nada for feito.

```python
# Retorna o valor pendente de pagamento sob o periodo abril a setembro de 2005
def valor_default_a_pagar(data, feature_default):
    return ((data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT6'] - 
                 data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT6']) + 
             (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT5'] - 
                  data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT5']) + 
             (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT4'] - 
                  data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT4']) +
             (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT3'] - 
                  data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT3']) +
             (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT2'] - 
                  data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT2']) +
             (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT1'] - 
                  data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT1'])).sum()
```

Sobre a taxa de inadimplência, trata-se do percentual dos créditos que a empresa possui que estão em atraso. 
Para calcular, é preciso somar todos os débitos em atraso e dividir pelo faturamento total previsto no período analisado.
(fontes: https://www.justonline.com.br/tudo-sobre-emprestimo/o-que-voce-precisa-saber-sobre-taxa-de-inadimplencia, https://meucrediario.com.br/blog/indice-de-inadimplencia-da-loja/)

São considerados valores em atraso com mais de 90 dias. Portanto, consideraremos para calcular essa taxa a soma das faturas sob o período (abril a junho) e o débito em atraso é considerado o valor da fatura dos clientes marcados como default sob esse mesmo período menos tudo o que ele já tiver pago até o fim de setembro.

```python
# Soma do valor pendente de abril a junho de 2005 (não pago até fim de setembro) 
# dos clientes DEFAULT_PAY tambem marcados em feature_default
def valor_pendente_default_inadimplencia(data, feature_default):
    
    fatura_default = (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT4']
                      + data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT5'] 
                      + data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT6']).sum()
    
    pagamento_default = (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT1'] 
                          + data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT2'] 
                          + data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT3'] 
                          + data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT4'] 
                          + data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT5']  
                          + data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT6']).sum()
    
    # Calculo do valor pendente
    return fatura_default - pagamento_default
```

```python
# Retorna a taxa de inadimplencia em porcentagem
def taxa_inadimplencia(data, feature_default, valor_pendente=None):
    
    # Soma das faturas de abril a junho de 2005
    faturas = (data['BILL_AMT4'] + data['BILL_AMT5'] + data['BILL_AMT6']).sum()
    
    if valor_pendente==None:
        valor_pendente = valor_pendente_default_inadimplencia(data, feature_default)
    
    return (valor_pendente/faturas) * 100
```

Se nada for feito, estamos com um prejuízo de mais de 750 milhões de reias, sendo que mais de 320 milhões de reais estão atrasados mais de três meses e temos uma taxa de inadimplência de 23.02%.

Vamos também analisar uma segunda forma de impacto, se considerado apenas o mês de setembro.

```python
# Retorna o valor pendente de pagamento respectivo somente ao mês de setembro
def valor_default_a_pagar_setembro(data, feature_default):
    return (data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'BILL_AMT1'] - 
                  data.loc[(data['DEFAULT_PAY'] == 1) & (data[feature_default] == 1), 'PAY_AMT1']).sum()
```

### Calculando impacto financeiro das possíveis regras em treino

Como dito na entrevista, quando o cliente paga a fatura integralmente, temos um lucro de aproximadamente 5% do valor da fatura. Se o cliente não paga, a gente perde todo o valor da fatura. 
Além disso, cada ligação custa R$50,00 independentemente do tempo.

```python
PRECO_LIGACAO = 50 # Cada ligação custa R$50,00 independentemente do tempo
```

### Tentativa 1 impacto financeiro

Para calcular o impacto financeiro da regra, vamos primeiro considerar que os clientes selecionados que eram default, após terem recebido a ligação, vão realizar o pagamento e teremos um lucro de 5% em cima do valor total pendente com mais de 3 meses de atraso.

```python
def impacto_financeiro_1(data, feature_default):
    # Calculando gasto em ligacao
    gasto_em_ligacao = data[data[feature_default]==1].shape[0] * PRECO_LIGACAO
    print(f"Teremos R${gasto_em_ligacao} de gasto em ligação")
    
    # Calculando valor pendente total 
    total_valor_pendente = valor_default_a_pagar(data, 'DEFAULT_PAY')
    print(f"Todos os clientes default na base estavam devendo um valor de R${total_valor_pendente:.2f}.") 
    # Inadimplencia de todos os clientes
    total_valor_pendente_inadimplencia = valor_pendente_default_inadimplencia(data, 'DEFAULT_PAY')
    print(f"Sendo que R${total_valor_pendente_inadimplencia:.2f} desse valor está atrasado mais de 3 meses.\n") 
    
    # Calculando o valor pendente dos clientes default listados por feature_default
    valor_pendente_feature = valor_default_a_pagar(data, feature_default)
    print(f"Os clientes default escolhidos estavam devendo um valor de R${valor_pendente_feature:.2f}.") 
    # Inadimplencia dos clientes escolhidos
    valor_pendente_feature_inadimplencia = valor_pendente_default_inadimplencia(data, feature_default)
    print(f"Sendo que R${valor_pendente_feature_inadimplencia:.2f} desse valor está atrasado mais de 3 meses.\n") 
    
    # Consideramos que todos os clientes que ligamos, paguem as faturas atrasadas mais de três meses
    valor_a_recuperar = valor_pendente_feature_inadimplencia
    
    # Calculando o valor recuperado
    valor_recuperado = valor_a_recuperar*1.05 - gasto_em_ligacao
    valor_sem_recuperacao = total_valor_pendente - valor_recuperado
    valor_sem_recuperacao_inadimplencia = total_valor_pendente_inadimplencia - valor_recuperado
    print(f"Portanto, seria possível recuperar R${valor_recuperado:.2f} do valor pendente.")
    if valor_sem_recuperacao < 0 :
        print(f"Obtendo um lucro de R${abs(valor_sem_recuperacao):.2f}.")
    else:
        print(f"Diminuindo o prejuízo para R${valor_sem_recuperacao:.2f}.")
    print(f"Sendo que R${valor_sem_recuperacao_inadimplencia:.2f} desse valor está atrasado mais de 3 meses.")
    
    # Calculando a taxa de inadimplencia
    val = (total_valor_pendente_inadimplencia - valor_a_recuperar) if valor_a_recuperar >= 0 else 0
    print(f"Nesse caso, temos que a taxa de inadimplencia é de {taxa_inadimplencia(data, feature_default, val):.2f}%.") 
```

```python
print("Regra 1 - Ligar para quem atrasou o pagamento no último mês.\n")
dados_regra['DEFAULT_REGRA'] = default_regra1.astype('int64')
impacto_financeiro_1(dados_regra, 'DEFAULT_REGRA')
```

### Decision tree

Vamos começar com uma simples árvore de decisão. 

Para preparar nossos folds para o cross-validation, vamos utilizar o Stratified K Fold, que garante que cada classe seja bem representada em cada fold.

```python
# Setando parâmetros
N_JOBS = 4 #number of parallel jobs 
RANDOM_STATE = 2020
```

Dependendo da escolha dos parâmetros, temos um resultado diferente. Portanto, utilizaremos o RandomizedSearchCV para realizar testes nas configurações para escolher a combinação que performa melhor no cross-validation.

```python
param_grid = {'max_depth': np.arange(2,10),
              'min_samples_split' : np.arange(2,20,2),
              'min_samples_leaf' : np.arange(1,21,2),
             'random_state': [2020]}

#create a grid
grid_tree = RandomizedSearchCV(estimator = DecisionTreeClassifier(), param_distributions = param_grid, 
                                n_iter = 50, cv = skf, random_state=RANDOM_STATE, n_jobs = N_JOBS,
                                scoring = 'roc_auc')

#training
%time grid_tree.fit(X_res_train, y_res_train)

#let's see the best estimator
best_tree = grid_tree.best_estimator_
print(best_tree)
print("_"*40)
#with its score
print("Melhor pontuação com cross-validation {}%".format(round(grid_tree.best_score_ * 100,3)))
#score on test
prediction = best_tree.predict(X_test)
print("Score do teste: {}%".format(round(accuracy_score(y_true = y_test, y_pred = prediction) * 100,3)))
```

#### Random Forrest Grid Search

É interessante testar alguns diferentes parâmetros pois eles podem alterar o resultado. Precisamos nos atentar para não ocasionar overfitting. Temos que:

* Quanto maior o max_depth, maior a acurácia em treino porém diminui um pouco em teste.
* Quanto maior o n_estimators, maior a acurácia tanto em treino quanto em teste, porém a partir de um certo ponto a melhor é irrisória, se tornando estável.
* Quanto maior o min_samples_leaf e o min_samples_split, menor a acurácia em treino, porém em teste é praticamente estável, tendo uma leve melhora.

Vamos então utilizar realizar uma busca sob um conjunto de parâmetros para encontrar a melhor combinação.

```python
# Número de árvores no random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 10)]

# Número de features para considerar em cada divisão
max_features = ['auto', 'sqrt', 'log2']

# Número máximo de níveis na árvore
max_depth = [int(x) for x in np.linspace(2, 30, num = 15)]
max_depth.append(None)

# Número mínimo de amostras necessárias em cada nó folha 
min_samples_leaf = [int(x) for x in np.linspace(2, 50, num = 25)]

# Número mínimo de amostras para dividir um nó interno 
min_samples_split = [int(x) for x in np.linspace(2, 100, num = 50)]

# Criar o random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_leaf': min_samples_leaf,
               'min_samples_split' : min_samples_split,
               'n_jobs' : [N_JOBS],
               'random_state' : [RANDOM_STATE]}
```

Isso significa que iríamos cultivar 1.000.000 de florestas, cerca de 100 milhões de árvores. Portanto, realizaremos uma busca aleatória, na qual uma certa quantidade de pontos são escolhidos aleatoriamente neste espaço de parâmetros e encontra o melhor estimador entre eles (apenas 50 para nós devido a acelerar a pesquisa). 

Vamos utilizar como scoring 'roc_auc' e utilizaremos o mesmo StratifiedKFold utilizado na Decision Tree no cross-validation.

```python
grid_forest = RandomizedSearchCV(estimator = RandomForestClassifier(), param_distributions = random_grid, 
                                n_iter = 50, cv = skf, random_state=RANDOM_STATE, n_jobs = N_JOBS,
                                scoring = 'roc_auc')

# Treinando
%time grid_forest.fit(X_res_train, y_res_train)

# Vamos ver o melhor estimador
best_forest = grid_forest.best_estimator_
print(best_forest)
print("_"*40)
# Com seu score
print("Melhor pontuação com cross-validation: {}%".format(round(grid_forest.best_score_ * 100,3)))
# Score do teste
prediction = best_forest.predict(X_test)
print("Score do teste: {}%".format(round(accuracy_score(y_true = y_test, y_pred = prediction) * 100,3)))
```

Vamos visualizar a importância das features.

Temos como variáveis mais importantes PAY_0, PAY_2, PAY_3, PAY_4, LIMIT_BAL e PAY_AMT1.
